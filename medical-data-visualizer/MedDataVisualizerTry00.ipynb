```python
med_long_df = pd.melt(med_df, id_vars = ['cardio'], value_vars = ['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'], var_name= 'variable', value_name='value')
```

```python
plot = sns.catplot(data = med_long_df, x = 'variable', hue = 'value', col = 'cardio', kind = 'count', palette = 'Set2')
plot.set_ylabels('total')
plot.savefig('figura.PNG')
plt.show()
```

```python
med_df[(med_df.height <= med_df.height.quantile(0.975)) & (med_df.height >= med_df.height.quantile(0.025)) & (med_df.weight <= med_df.weight.quantile(0.975)) & (med_df.weight >= med_df.weight.quantile(0.025)) & (med_df.ap_lo <= med_df.ap_hi)]
```

```python
med_df_hot = med_df[(med_df.height <= med_df.height.quantile(0.975)) & (med_df.height >= med_df.height.quantile(0.025)) & (med_df.weight <= med_df.weight.quantile(0.975)) & (med_df.weight >= med_df.weight.quantile(0.025)) &(med_df.ap_lo <= med_df.ap_hi)]
med_df_corr = med_df_hot.corr()
mask_med = np.triu(np.ones_like(med_df_corr, dtype=bool))
ax = sns.heatmap(med_df_corr, annot = True, fmt = '.1f', mask = mask_med, linewidths=.5);
fig = plt.gcf()
fig.savefig('headmap.PNG');
```
