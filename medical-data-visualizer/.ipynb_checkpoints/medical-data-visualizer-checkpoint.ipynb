```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
df = pd.read_csv('medical_examination.csv')
df.loc[df['weight'] / np.square(df['height'] / 100) > 25, 'overweight'] = 1
df.loc[df['weight'] / np.square(df['height'] / 100) <= 25, 'overweight'] = 0
df['overweight'] = df['overweight'].astype('int64')
df.loc[df['cholesterol'] == 1, 'cholesterol'] = 0
df.loc[df['cholesterol'] >= 2, 'cholesterol'] = 1
df.loc[df['gluc'] == 1, 'gluc'] = 0
df.loc[df['gluc'] >= 2, 'gluc'] = 1
df_cat = pd.melt(df, id_vars = ['cardio'], value_vars = ['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'], var_name= 'variable', value_name='value')
cat = sns.catplot(data = df_cat, x = 'variable', hue = 'value', col = 'cardio', kind = 'count', palette = 'Set2')
fig = plt.gcf()
ax = plt.gca()
cat.set_ylabels('total')
fig.savefig('catplot.png')
```

```python
df_heat = df[(df.height <= df.height.quantile(0.975)) & (df.height >= df.height.quantile(0.025)) & (df.weight <= df.weight.quantile(0.975)) & (df.weight >= df.weight.quantile(0.025)) &(df.ap_lo <= df.ap_hi)]
corr = df_heat.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
fig, ax = plt.subplots(figsize = (10,10))
sns.heatmap(corr, annot = True, fmt = '.1f', mask = mask, linewidths=.5 )
#fig = plt.gcf()
#ax = plt.gca()
fig.savefig('heatmap.png')
plt.plot()
```
