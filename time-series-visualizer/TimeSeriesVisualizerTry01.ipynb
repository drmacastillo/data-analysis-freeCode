```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para el DataFrame
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', end='2023-12-31', freq='D')
values = np.random.randint(1, 3001, size=len(dates))

# Crear el DataFrame con fechas como índice
df = pd.DataFrame({'c2': values}, index=dates)

# Calcular el promedio mensual de 'c2' por año y mes
df_monthly_avg = df.resample('ME').mean()

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
years = df_monthly_avg.index.year.unique()
bar_width = 0.8 / 12

for i, month in enumerate(range(1, 13)): ### i va de 0 a 11 y month va de 1 a 12
    monthly_data = df_monthly_avg[df_monthly_avg.index.month == month] ### todos los datos de un mismo mes, comenzando por el mes 1 hasta el mes 12
    plt.bar(monthly_data.index.year + bar_width * i, monthly_data['c2'], width=bar_width, align='center', label=f'Mes {month}')

plt.xlabel('Año')
plt.ylabel('Promedio de c2')
plt.title('Promedio Mensual de c2 por Año')
plt.xticks(years)
plt.legend(title='Mes')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from calendar import month_name

# Crear datos para el DataFrame
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', end='2023-12-31', freq='M')
values = np.random.randint(1, 3001, size=len(dates))

# Crear el DataFrame con fechas como índice
df = pd.DataFrame({'c2': values}, index=dates)

# Calcular el promedio mensual de 'c2' por año y mes
df_monthly_avg = df.resample('M').mean()

# Crear la gráfica de barras con línea promedio
plt.figure(figsize=(12, 6))
years = df_monthly_avg.index.year.unique()
bar_width = 0.8 / 12

for i, month in enumerate(range(1, 13)):
    monthly_data = df_monthly_avg[df_monthly_avg.index.month == month]
    month_name_spanish = month_name[month]
    plt.bar(monthly_data.index.year + 0.4 + bar_width * i, monthly_data['c2'], width=bar_width, align='center', label=month_name_spanish)

# Calcular y graficar la línea promedio por año
yearly_avg = df.resample('Y').mean()
plt.plot(yearly_avg.index.year, yearly_avg['c2'], color='red', marker='o', label='Promedio Anual')

plt.xlabel('Año')
plt.ylabel('Promedio de c2')
plt.title('Promedio Mensual de c2 por Año con Línea Promedio Anual')
plt.xticks(years)
plt.legend(title='Mes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Crear un DataFrame de ejemplo con fechas y valores aleatorios
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100)
df = pd.DataFrame({'Date': dates, 'Value': values})
df['Month'] = df['Date'].dt.month_name()
sns.boxplot(x='Month', y='Value', data=df)

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```
